```python
import requests
from bs4 import BeautifulSoup
```

```python
url ='https://wink.ru/collections/normalnye-novinki'

r = requests.get(url)
r.text
```

```python
soup = BeautifulSoup(r.text, 'lxml')
soup
```

```python
soup.find_all('div',{'class':'r11d8112 rnj3w1t r1yh4fwl'})
```

```python
new_film = soup.find_all('div',{'class':'r11d8112 rnj3w1t r1yh4fwl'})[0]
```

```python
names = []
for t in new_film.find_all('h4', {'class':'r1lbxtse t1j371zb rjqy0lg'}):
    names.append(t.text)
```

```python
names
```

```python
year = []
for t in new_film.find_all('span', {'data-test':'mic-year'}):
    year.append(t.text)
```

```python
year
```

```python
time = []
for t in new_film.find_all('span', {'data-test':'mic-duration'}):
    time.append(t.text)
time
```

```python
import pandas as pd
df = pd.DataFrame({'name': names, 'year': year, 'time':time})
df.head() #создаем таблицу
```
